```python
import pandas as pd
import numpy as np
```

```python
crimes = pd.read_excel('../Data/Cleaned/crimes_district.xlsx')
```

```python
crimes.head()
```

```python
totalcrimes = crimes.iloc[:,5:].sum().sort_values(ascending=False)
```

```python
totalcrimes = pd.DataFrame(totalcrimes)
```

```python
totalcrimes = totalcrimes.rename(columns={0:'total_crimes'})
```

```python
totalcrimes = totalcrimes.iloc[1:,:]
```

```python
totalcrimes
```

```python
totalcrimes['total_crimes_p'] = totalcrimes['total_crimes']/totalcrimes['total_crimes'].sum()
```

```python
totalcrimes['total_crimes_p'] = totalcrimes[['total_crimes_p']].round(2)
```
