### 2.Reading Exploring and Cleaning Data

```python
#!ls ../Data/cleaned
```

```python
crimesdata = pd.read_excel('../Data/cleaned/crimesdata_clean.xlsx')
crimesdata.head()
```

```python
crimesdata.shape
```

```python
crimesdata = crimesdata.drop(columns=['Zipcode','Location'])
```

```python
crimesdata['Total_crimes'] = crimesdata.iloc[:,2:].sum(axis=1)
```

```python
crimesdata.head()
```

```python
crimes_district = crimesdata.groupby(['Year','District']).sum().reset_index()
```

```python
crimes_district.head()
```
