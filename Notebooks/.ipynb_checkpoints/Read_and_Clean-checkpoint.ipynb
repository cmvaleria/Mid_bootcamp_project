## Berlin Crimes HZ (population factor) Cleaning

```python
#ls ../Data/Raw/WHAT
```

```python
crimesberlin_hz = pd.read_excel('../Data/Raw/WHAT/Berlin_crimes_official_HZ.xlsx')
crimesberlin_hz.head()
```

```python
crimesberlin_hz.isnull().sum()
```

```python
crimesberlin_hz = crimesberlin_hz.dropna()
```

```python
crimesberlin_hz.info()
```

```python
crimesberlin_hz.Robbery.unique()
```

```python
crimesberlin_hz[crimesberlin_hz.District == 'Mitte'].head(11)
```

```python
crimesberlin_hz['Robbery'] = crimesberlin_hz['Robbery'].astype(int)
```

```python
crimesberlin_hz.info()
```

```python
crimesberlin_hz['Street_robbery'] = crimesberlin_hz['Street_robbery'].astype(int)
crimesberlin_hz['Body_injuries'] = crimesberlin_hz['Body_injuries'].astype(int)
crimesberlin_hz['Agg_assoult'] = crimesberlin_hz['Agg_assoult'].astype(int)
crimesberlin_hz['Threat'] = crimesberlin_hz['Threat'].astype(int)
crimesberlin_hz['Theft'] = crimesberlin_hz['Theft'].astype(int)
crimesberlin_hz['Vehicle'] = crimesberlin_hz['Vehicle'].astype(int)
crimesberlin_hz['From_car'] = crimesberlin_hz['From_car'].astype(int)
crimesberlin_hz['Bike'] = crimesberlin_hz['Bike'].astype(int)
crimesberlin_hz['Housebreaking'] = crimesberlin_hz['Housebreaking'].astype(int)
crimesberlin_hz['Fire'] = crimesberlin_hz['Fire'].astype(int)
crimesberlin_hz['Damage_property'] = crimesberlin_hz['Damage_property'].astype(int)
crimesberlin_hz['Graffiti'] = crimesberlin_hz['Graffiti'].astype(int)
crimesberlin_hz['Drugs'] = crimesberlin_hz['Drugs'].astype(int)
#crimesberlin_hz['Domestic'] = crimesberlin_hz['Domestic'].astype(int)

crimesberlin_hz.info()
```

```python
crimesberlin_hz.columns
#crimesberlin_hz['Domestic'].unique()
```

```python
crimesberlin_hz = crimesberlin_hz.rename(columns={"Domestic ": "Domestic"})
```

```python
crimesberlin_hz.columns
```

```python
crimesberlin_hz['Domestic'] = crimesberlin_hz['Domestic'].astype(int)
```

```python
crimesberlin_hz.info()
```
