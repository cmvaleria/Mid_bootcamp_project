```python
crimesdata_hz = pd.read_excel('../Data/cleaned/crimesberlin_hz_clean.xlsx')
crimesdata_hz.head()
```

```python
crimesdata_hz['Total_crimes'] = crimesdata_hz.iloc[:,4:].sum(axis=1)
```

```python
crimesdata_hz = crimesdata_hz.drop(columns=['Zipcode','Location'])
```

```python
crimesdata_hz.head()
```

```python
crimes_district_hz = crimesdata_hz.groupby(['Year','District']).sum().reset_index()
```

```python
crimes_district_hz.head()
```

```python
# crimes_district_hz.to_excel('../Data/cleaned/crimes_district_hz.xlsx', index = False)
```

```python
crimes_2021 = crimes_district_hz[crimes_district_hz['Year']==2021]
```

```python
crimes_2021.head()
```

```python
crimes_2021.shape
```

```python
crimes_2021 = crimes_2021.rename(columns={"Total_crimes": "Total_crimes_Hz"})
```

```python
crimes_2021 = crimes_2021[['Year','District','Total_crimes_Hz']]
```

```python
crimes_2021
```

```python
def district_id (x):
    if x['District'] == 'Mitte':
        return 1
    elif x['District'] == 'Friedrichshain-Kreuzberg':
        return 2
    elif x['District'] == 'Pankow':
        return 3
    elif x['District'] == 'Charlottenburg Nord':
        return 4
    elif x['District'] == 'Spandau':
        return 5
    elif x['District'] == 'Steglitz-Zehlendorf':
        return 6
    elif x['District'] == 'Tempelhof-Schöneberg':
        return 7
    elif x['District'] == 'Neukölln':
        return 8
    elif x['District'] == 'Treptow-Köpenick':
        return 9
    elif x['District'] == 'Marzahn-Hellersdorf':
        return 10
    elif x['District'] == 'Lichtenberg':
        return 11
    elif x['District'] == 'Reinickendorf':
        return 12
```

```python
crimes_2021['district_id'] =  crimes_2021.apply(district_id, axis=1)
```

```python
crimes_2021.head()
```

```python
crimes_district = pd.read_excel('../Data/cleaned/crimes_district.xlsx')
crimes_district.head()
```

```python
crimes_district = crimes_district[crimes_district['Year']==2021]
```

```python
crimes_district['district_id'] =  crimes_district.apply(district_id, axis=1)
```

```python
crimes_district.head()
```

```python
crimes_district.columns
```

```python
crimes_district = crimes_district[[ 'district_id','Year', 'District', 'Robbery', 'Street_robbery', 'Body_injuries',
       'Agg_assoult', 'Threat', 'Theft', 'Vehicle', 'From_car', 'Bike',
       'Housebreaking', 'Fire', 'Damage_property', 'Graffiti', 'Drugs',
       'Domestic ', 'Total_crimes']]
```

```python
crimes_district.sort_values(by='district_id', ascending=True)
```

```python
crimes_district.merge(crimes_2021, how='right', on='district_id')
```

```python
crimes_2021 = crimes_2021.sort_values(by='district_id', ascending=True)
```

```python
crimes_2021
```

```python
crimes_2021[['Total_crimes_Hz']]
```

```python
crimes_2021 = pd.concat([crimes_district, crimes_2021[['Total_crimes_Hz']]], axis=1).sort_values(by='district_id', ascending=True)
```

```python
crimes_2021.head()
```

```python
crimes_2021 = crimes_2021.reset_index(drop=True)
```
